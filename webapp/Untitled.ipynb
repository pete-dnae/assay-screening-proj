```python
import urllib.request
import json
from pandas.io.json import json_normalize
```

```python
response = urllib.request.urlopen("http://assay-screening.herokuapp.com/api/reagents/")
byteString = response.read()
json_data = json.loads(byteString)
```

```python
dataFrame = json_normalize(json_data)
```

```python
import pandas as pd 
```

```python
metaData = pd.read_excel('C:/Users/Prasanna.Gnanaraj/Documents/20180122 target_length.xlsx')
```

```python
def find_amplicon_length(target):
    amplicon_length = metaData[metaData.target==target]['length'].values[0]
    
    return int(amplicon_length)
    
```

```python
dataFrame['opaque_json_payload'] = dataFrame[dataFrame.category=='assay']['name'].apply(find_amplicon_length)
```

```python
dataFrame['opaque_json_payload']=dataFrame['opaque_json_payload'].fillna('null')
```

```python
dataFrame.to_csv(r'C:\Users\Prasanna.Gnanaraj\Documents\assay-screening\app\model_builders\customers-seed-data\reagents.csv')
```
